```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/Lenovo/OneDrive/Desktop/zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'], axis=1)
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if (value == 'NEW' or value == '-'):
        return np.nan
    else:
        return float(str(value).split('/')[0])

df['rate'] = df['rate'].apply(handlerate)
df.head()
```

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df.rate.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2Plates', 'listed_in(type)':'Type', 'listed_in(city)':'City'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['City'].unique()
```

```python
df.drop(['City'], inplace = True, axis = 1)
```

```python
df.head()
```

```python
df['Cost2Plates'].unique()
```

```python
df['Cost2Plates'] = df['Cost2Plates'].str.replace(',', '').astype(float)
```

```python
df.head()
```

```python
df['Cost2Plates'].unique()
```

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
location = df['location'].value_counts(ascending  = False)
location
```

```python


location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
